```python
class myappclass:
    def __init__(self):
        self.window=Tk()
         #=========Background Image=============
        
        w = self.window.winfo_screenwidth()
        h = self.window.winfo_screenheight()
        
        self.bimg1 = Image.open("image//img2.jpeg")
        self.bimg1 = self.bimg1.resize((w,h), Image.LANCZOS)

        self.bimg2 = ImageTk.PhotoImage(self.bimg1)

        self.bklb1=Label(self.window,image=self.bimg2)
        self.bklb1.place(x=0,y=0)
        
        self.window.title("Loan Prediction App")
        self.window.geometry("%dx%d+%d+%d"%(500,550,500,100))
        self.headlbl = Label(self.window, text="Loan Prediction", font=("Lucida Calligraphy", 20))
        self.l1 = Label(self.window, text=" Gender")
        self.l2 = Label(self.window, text=" Are you  Married ")
        self.l3 = Label(self.window, text="How much persons depends on you")
        self.l4 = Label(self.window, text="What is your Education Qualification")
        self.l5 = Label(self.window, text="Are you Self Employed ")
        self.l6 = Label(self.window, text="How much is your income")
        self.l7 = Label(self.window, text="How much Coapplicant's Income")
        self.l8 = Label(self.window, text="Loan Amount (in thousand)")
        self.l9 = Label(self.window, text=" Loan Amount Term (in days)")
        self.l10 = Label(self.window, text="Credit History ")
        self.l11 = Label(self.window, text="Property Area") 
        self.l12 = Label(self.window, text=" -------------")
        
      #************************************************************* 
        self.v1 = StringVar()
        self.r1 = Radiobutton(self.window, text='Male', value='Male',  variable=self.v1)
        self.r2 = Radiobutton(self.window, text='Female', value='Female', variable=self.v1)
        self.v1.set("  ")
        
        self.list1=["Yes","No"]
        self.v2 = StringVar() 
        self.c1= Combobox(self.window, values=self.list1, textvariable=self.v2, state='readonly')
        self.v2.set("Yes")
        
        self.t1 = Entry(self.window)
        
        self.list2=["Graduate","Not Graduate"]
        self.v3 = StringVar() 
        self.c2= Combobox(self.window, values=self.list2, textvariable=self.v3, state='readonly')
        self.v3.set("Graduate")
        
        self.v4 = StringVar() 
        self.c3= Combobox(self.window, values=['Yes',"No"], textvariable=self.v4, state='readonly')
        self.v4.set("Yes")
        
        
        self.t2 = Entry(self.window)
        
        self.t3 = Entry(self.window)
        
        self.t4 = Entry(self.window)
        
        self.t5 = Entry(self.window)
        
        self.v5 = StringVar() 
        self.c4= Combobox(self.window, values=['Yes',"No"], textvariable=self.v5, state='readonly')
        self.v5.set("Yes")
        
        self.list3=["Urban","Semi-Urban","Rural"]
        self.v6 = StringVar() 
        self.c5= Combobox(self.window, values=self.list3, textvariable=self.v6, state='readonly')
        self.v6.set("Urban")
        self.btn1 =Button(self.window, text="Predict Loan Approval",background='silver',command=self.predict)
        
 #***************************************************************************************************************************
        self.headlbl.place(x=40, y=20)
        x1 = 40
        y1 = 100
        self.l1.place(x=x1, y=y1)
        self.r1.place(x=x1 + 250, y=y1)
        self.r2.place(x=x1 + 250+100, y=y1)
        y1 += 50
        self.l2.place(x=x1, y=y1)
        self.c1.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l3.place(x=x1, y=y1)
        self.t1.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l4.place(x=x1, y=y1)
        self.c2.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l5.place(x=x1, y=y1)
        self.c3.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l6.place(x=x1, y=y1)
        self.t2.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l7.place(x=x1, y=y1)
        self.t3.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l8.place(x=x1, y=y1)
        self.t4.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l9.place(x=x1, y=y1)
        self.t5.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l10.place(x=x1 ,y=y1)
        self.c4.place(x=x1 + 250, y=y1)
        y1 += 50
        self.l11.place(x=x1, y=y1)
        self.c5.place(x=x1 + 250, y=y1)
        
        self.btn1.place(x=x1+100, y=y1+70)
        y1 += 50
        
        self.l12.place(x=x1+100, y=y1+80)
        y1 += 50
        
        self.window.mainloop()
        
    def predict(self):
        Gender=self.v1.get()
        if(Gender=='Male'):
            Gender=1
        else: # Female
            Gender=0
                
        Married=self.v2.get()
        if(Married=='Yes'):
            Married=1
        else: 
            Married=0
                
        Dependent_persons=int(self.t1.get()) 
            
        Qualification=self.v3.get()
        if(Qualification=='Graduate'):
            Qualification=1
        else: #not graduate
            Qualification=0
                
        Self_employed=self.v4.get()
        if(Self_employed=='Yes'):
            Self_employed=1
        else:#Not Employed 
            Self_employed=0 
                
        Income=int(self.t2.get())
            
        Coapplicant_income=float(self.t3.get())
            
        Loan_amount=float(self.t4.get())
            
        Loan_amount_term=float(self.t5.get())
            
        Credit_history=self.v5.get()
        if(Credit_history=='Yes'):
            Credit_history=1
        else: 
            Credit_history=0
                
        Property_area=self.v6.get()
        if(Property_area=='Urban'):
            Property_area=1
            Property_area_Semi_Urban=0
            Property_area_Rural=0         
        elif(Property_area=='Semi-Urban'):
            Property_area_Urban=0
            Property_area=1
            Property_area_Semi_Urban=0
        else: # Rural
            Property_area=1
            Property_area_Urban=0
            Property_area_Semi_Urban=0
                             
        prediction=m.predict([[Gender,Married,Qualification,Dependent_persons,Self_employed,Income,
                                   Coapplicant_income,Loan_amount,Loan_amount_term,Credit_history,Property_area]]) 
        output=round(prediction[0],2)
        if output==0:
            self.l12.config(text="SORRY!! YOU ARE NOT ELIGIBLE FOR TAKING LOAN")
        else:
            self.l12.config(text="YOU CAN TAKE LOAN")

     
```
